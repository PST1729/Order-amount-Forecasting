```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, y='amount_in_usd')
plt.title('distribution amount_in_usd')
plt.ylabel('amount_in_usd')
plt.show()
```

```python
outliers_amount = df['amount_in_usd'].values[df['amount_in_usd'] > df['amount_in_usd'].quantile(0.75) + 1.5 * (df['amount_in_usd'].quantile(0.75) - df['amount_in_usd'].quantile(0.25))]
len(outliers_amount)
```

```python
df['amount_in_usd'].describe()
```

```python
df['amount_in_usd'].shape
```

```python
Q1 = df['amount_in_usd'].quantile(0.25)
Q3 = df['amount_in_usd'].quantile(0.75)
IQR = Q3 - Q1

outlier_threshold = 1.5

median_value = df['amount_in_usd'].median()
outliers = df[(df['amount_in_usd'] < Q1 - outlier_threshold * IQR) | (df['amount_in_usd'] > Q3 + outlier_threshold * IQR)]
df.loc[outliers.index, 'amount_in_usd'] = median_value

total_outliers = len(outliers)

print("DataFrame with outliers replaced:")
print(df)
print("\nTotal number of outliers:", total_outliers)
```

# LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
selected_columns = selected_columns = ['SALES_ORG','COMPANY_CODE','CREDIT_STATUS','CUSTOMER_NUMBER','unique_cust_id','DISTRIBUTION_CHANNEL_encoded',
'DIVISION_encoded','ORDER_AMOUNT_LOG']
X = df[selected_columns]
y = df['amount_in_usd_log']

# Convert YearMonth to year and month columns
X['Year'] = df['YearMonth'].dt.year
X['Month'] = df['YearMonth'].dt.month

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print(y_pred)
```

```python
predictions_log = model.predict(X_test) 
predictions = np.exp(predictions_log)
print(predictions)
```

Perform Regression model evaluations like MSE, RMSE, R-Square etc.

```python
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("Mean Absolute Error (MAE):", mae)
print("R-squared (R^2):", r2)
```

 Checking the overfitting or underfitting of the model

```python
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
# Calculate the mean squared error (MSE) on training set and test set
mse_train = mean_squared_error(y_train, y_train_pred)
mse_test = mean_squared_error(y_test, y_test_pred)

print('MSE on training set:', mse_train)
print('MSE on test set:', mse_test)
```
